```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
proc = pd.read_csv('analyze_multi-process.csv')
thread = pd.read_csv('analyze_multi-threaded.csv')
single = pd.read_csv('analyze_single.csv')
```

```python
proc_med = proc.groupby('num_clients').median()
thread_med = thread.groupby('num_clients').median()
single_med = single.groupby('num_clients').median()
```

```python
display(proc_med), display(thread_med), display(single_med)
```

```python
plt.plot(proc_med.index, np.log10(proc_med['total_time']), label='multi-process')
plt.plot(thread_med.index, np.log10(thread_med['total_time']), label='multi-threaded')
plt.plot(single_med.index, np.log10(single_med['total_time']), label='single-threaded')
plt.xlabel('Number of clients')
plt.ylabel('Time (s) [log scale]')
plt.title('Median total time vs different numbers of clients')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
plt.plot(proc_med.index, proc_med['total_time'], label='multi-process')
plt.plot(thread_med.index, thread_med['total_time'], label='multi-threaded')
plt.xlabel('Number of clients')
plt.ylabel('Time (s) [standard scale]')
plt.title('Median total time vs different numbers of clients')
plt.grid(alpha=0.5)
plt.legend()
plt.show()
```
